**딕셔너리에서 Key는 고유한 값**이므로 **중복되는 Key 값을 설정해 놓으면 하나를 제외한 나머지 것들이 모두 무시**된다는 점을 주의해야 한다. 다음 예에서 볼 수 있듯이 동일한 Key가 2개 존재할 경우 1:'a'라는 쌍이 무시된다.

```python
a = {1:'a', 1:'b'}

a
```

**리스트는 그 값이 변할 수 있기 때문에 Key로 사용할 수 없다**. 하지만 값이 변하지 않는 튜플은 Key로 쓸 수 있다. 아래 예처럼 리스트를 Key로 설정하면 리스트를 키 값으로 사용할 수 없다는 오류가 발생한다.

```python
a = {[1,2] : 'hi'}  # 리스트 형태는 키 값으로 사용할 수 없다.
```

```python
a = {(1,2) : 'hi'}  #튜플은 괜찮다. 튜플의 요소값은 바꿀 수 없다. 에러남. 바뀌지않으면 키로 활용가능. 리스트는 요소 값을 바꿀수 있음
```

### 예시 코드

```python
A = [10, 60, 30, 56, 90, 77, 89, 40, 70]

total = 0
n = 0
while A:  # A 안에 원소가 있으면 항상 true 없으면 false
    score = A.pop()     # pop 이라하면 맨 끝에 원소가 튀어나옴. 70 이 들어옴. 그 다음 40
    if score >= 60:
        n += 1
        total += score

avg = total/n
print("총합:", total)
print("평균:", avg)

# 프로그램에서는 0 이면 false, 1 이상이면 true
```

### 사용자 정의함수

```python
# skip 함

def find_max (a,b): 
    if a > b : 
        y = a 
    else : 
        y = b 

    return y


# main
m = find_max(10,20) # 10, 20 : 인수(arguments) 
n = find_max(7,3) # 7, 3 : 인수 (arguments) 

print(m,n)
```

### 함수의 인수 값과 반환 값

```python
def calculate(kor, eng, math): 
    total = kor+ eng+ math 
    average = total / 3
    
    return int(average)

# main 

kor_score= int(input('국어성적을입력하시오: ')) 
eng_score= int(input('영어성적을입력하시오: ')) 
math_score= int(input('수학성적을입력하시오: '))

avg= calculate(kor_score, eng_score, math_score) 

print('평균:', avg)
```

### 반환값이 여러 개인 경우(사실은 여러 개가 아니라 하나의 튜플)

```python
def add_multiply(x,y): 
    sum = x + y 
    mult= x * y 
    
    return sum, mult # 튜플로반환한다.
    
# main 

a = int(input('Enter a : ')) 
b = int(input('Enter b : ')) 

t = add_multiply(a,b) 
print("반환값: ", t)
print()
m,n= add_multiply(a,b) # m 과 n 을 따로 받을 수도 있다.
print(m,n)
```

```python
# 입력 값의 개수가 달라질때
# 여러 개의 non-keyword arguments:*args 입력값들을 전부 모아 하나의 튜플로 만들기 때문에 입력값이 몇개이든 상관 없음
def friends(*args): 
    for n in args : 
        print(n.title())  # 단어의 첫글자는 대문자, 나머지는 소문자로 변환

# main

friends('alice', 'paul') 
print('------')   
print('-'*8)
friends('cindy', 'sally', 'david', 'tom')


print()

def friends_all(*args): # 튜플로 만들어줌
    print(args)
    
friends_all('cindy', 'sally', 'david')
```

```python
def score(*scores):
    i = 0
    total = 0
    for score in scores:
        if score > 60:
            total +=score
            i+=1
    
    return total, i
            
# main

total, students = score(50, 40, 70, 90, 100)

print('총합계: {0}, 학생 수: {1}'.format(total, students))
```
